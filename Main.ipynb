```python
import pandas as pd
# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/alpie13-blip/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv'
try:
    # This is a new change

    df = pd.read_csv(github_raw_url)
    print("Successfully loaded data from GitHub!")
    display(df.head())
except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```

```python
#plot each df col graph
import matplotlib.pyplot as plt
df.plot(kind='bar')
plt.show()
```

```python
#plot barchart for Brand col
import matplotlib.pyplot as plt
df['Brand'].value_counts().plot(kind='bar')
plt.show()
```

```python
#piechart for Brand col
import matplotlib.pyplot as plt
df['Brand'].value_counts().plot(kind='pie')
plt.show()
```

```python
#plot histogram for Price_SGD col with 10 bins
import matplotlib.pyplot as plt
df['Price_SGD'].hist(bins=10)
plt.show()

```

```python
#Plot boxplot for Price_SGD vs Brand
import matplotlib.pyplot as plt
df.boxplot(column='Price_SGD', by='Brand')
plt.show()
```
